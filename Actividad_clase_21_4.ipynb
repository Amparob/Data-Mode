##Transformación, join de datasets

```python

# Notas de Matemática y Lengua
df1 = pd.DataFrame({'Alumno': ['Alumno0', 'Alumno1', 'Alumno2'],
                    'Matermática': ['5', '6', '7'],
                    'Lengua': ['7', '7', '8']
                    })

# Notas de Historia y Geografía
df2 = pd.DataFrame({'Alumno': ['Alumno0', 'Alumno2', 'Alumno3'],
                    'Historia': ['4', '5', '4'],
                    'Geografía': ['6', '7', '9']
                    })

# Mostramos los dataframes
print(df1)
print('           ')
print(df2)

```

```python
# Hacemos join de los dataframes y mostramos las notas

merged_df = pd.merge(df1, df2, on='Alumno', how='left')
print(merged_df)
```

##Substitución de datos faltante por un tipo de dato default

```python
#Reemplazamos nulos por ceros

# Creando un DataFrame
df = pd.DataFrame({'columna1': [1, 2, None, 4],
                   'columna2': [5, None, None, 8]})

print(df)
```

```python
# Rellenando los valores nulos con 0

df = df.fillna(0) #Podemos reemplazar el cero por el valor que necesitemos
print(df)
```

##Eliminar columnas que no son necesarias

```python

# Creamos un DataFrame de ejemplo
data = {'nombre': ['Ana', 'Pedro', 'María'],
        'edad': [30, 25, 32],
        'ciudad': ['Madrid', 'Barcelona', 'Sevilla'],
        'correo': ['ana@gmail.com', 'pedro@gmail.com', 'maria@gmail.com'],
        'telefono': [123456789, 987654321, 567891234]}
df = pd.DataFrame(data)

print(df)
```

```python
# Eliminar las columnas 'correo' y 'telefono'
columnas_a_eliminar = ['correo', 'telefono']

#axis=1: Se refiere a las columnas.
#inplace=True: modifica el DataFrame original en lugar de crear una copia y realizar los cambios en ella
df.drop(columnas_a_eliminar, axis=1, inplace=True)

print(df)
```

##Agrupamiento

```python
# Creamos un DataFrame de ventas
data = {'Vendedor': ['Ana', 'Pedro', 'Ana', 'María', 'Pedro', 'Ana'],
        'Monto': [100, 200, 150, 300, 100, 50],
        'Operaciones': [1, 1, 2, 1, 3, 1]}
df_ventas = pd.DataFrame(data)

print("Datos originales:")
print(df_ventas)
```

```python
# Agrupamos por 'Vendedor' y sumamos sus montos
# El proceso es: 1. Agrupar, 2. Seleccionar columna, 3. Aplicar función (sum, mean, count)
ventas_totales = df_ventas.groupby('Vendedor')['Monto'].sum().reset_index()

print("\nVentas totales por vendedor:")
print(ventas_totales)
```

## Conversión de Mayúsculas

```python
# Creamos un DataFrame con nombres variados
data = {'Estudiante': ['ana gomez', 'PEDRO RODRIGUEZ', 'mAría loPez']}
df_nombres = pd.DataFrame(data)

print("Datos originales:")
print(df_nombres)
```

```python
# Convertir a MAYÚSCULAS
df_nombres['Estudiante_Mayus'] = df_nombres['Estudiante'].str.upper()
print(df_nombres)
```

```python
# Convertir a minúsculas
df_nombres['Estudiante_Minus'] = df_nombres['Estudiante'].str.lower()
print(df_nombres)
```

2. Transformación de Strings a Números: Convierte la cadena "99.5" a un tipo de dato flotante y luego a un tipo de dato entero. ¿Qué sucede con el valor decimal al hacer la conversión final?

```python
#Ejercicio 2
cadena_numero = "99.5"
print(f"Cadena original: {cadena_numero}, tipo: {type(cadena_numero)}")

# Convertir a flotante
numero_float = float(cadena_numero)
print(f"Convertido a float: {numero_float}, tipo: {type(numero_float)}")

# Convertir a entero
# Al convertir de float a int, la parte decimal se trunca (se elimina).
numero_int = int(numero_float)
print(f"Convertido a int: {numero_int}, tipo: {type(numero_int)}")

print("\nAl convertir de float a int, el valor decimal se trunca, es decir, se elimina. Por ejemplo, 99.5 se convierte en 99.")
```
